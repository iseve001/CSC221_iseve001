```python
class Money:
    def __init__(self, value):
        self.value = value
    def __str__(self):
        return f"${self.value:.2f}"
class Bills(Money):
    def __init__(self, denomination):
        value = Bills.get_value_from_denomination(denomination)
        super().__init__(value)
        self.denomination = denomination

    @staticmethod
    def get_value_from_denomination(denomination):
        denominations = {
            1: 1.00,
            5: 5.00,
            10: 10.00,
            20: 20.00,
        }
        return denominations.get(denomination, 0.00)

class One(Bills):
    def __init__(self):
        super().__init__(1)

class Five(Bills):
    def __init__(self):
        super().__init__(5)

class Ten(Bills):
    def __init__(self):
        super().__init__(10)

class Twenty(Bills):
    def __init__(self):
        super().__init__(20)

class Coins(Money):
    def __init__(self, denomination):
        value = Coins.get_value_from_denomination(denomination)
        super().__init__(value)
        self.denomination = denomination

    def get_value_from_denomination(denomination):
        denominations = {
            0.01: 0.01,
            0.05: 0.05,
            0.10: 0.10,
            0.25: 0.25,
        }
        return denominations.get(denomination, 0.00)


class Penny(Coins):
    def __init__(self):
        super().__init__(0.01)


class Nickel(Coins):
    def __init__(self):
        super().__init__(0.05)


class Dime(Coins):
    def __init__(self):
        super().__init__(0.10)


class Quarter(Coins):
    def __init__(self):
        super().__init__(0.25)


class Wallet:
    def __init__(self):
        self.money_list = []

    def add_money(self, money):
        self.money_list.append(money)

    def calculate_total(self):
        return sum(money.value for money in self.money_list)

    def __str__(self):
        return ", ".join(str(money) for money in self.money_list)


# Test cases
if __name__ == "__main__":
    wallet1 = Wallet()
    wallet2 = Wallet()
    wallet3 = Wallet()

    one = One()
    five = Five()
    ten = Ten()
    twenty = Twenty()
    penny = Penny()
    nickel = Nickel()
    dime = Dime()
    quarter = Quarter()
    wallet1.add_money(one)
    wallet1.add_money(five)
    wallet1.add_money(penny)
    wallet1.add_money(nickel)

    wallet2.add_money(twenty)
    wallet2.add_money(quarter)

    wallet3.add_money(ten)
    wallet3.add_money(dime)
    wallet3.add_money(dime)

    print(f"Wallet 1: {wallet1}")
    print(f"Total in Wallet 1: ${wallet1.calculate_total():.2f}")

    print(f"Wallet 2: {wallet2}")
    print(f"Total in Wallet 2: ${wallet2.calculate_total():.2f}")

    print(f"Wallet 3: {wallet3}")
    print(f"Total in Wallet 3: ${wallet3.calculate_total():.2f}")
```
